```python
import pandas as pd
```

```python
df = pd.read_json('https://data.cityofnewyork.us/resource/vfnx-vebw.json')
```

```python
display(df)
```

X and Y are latitude and longitude points, respectively. I want to create a map using these, so I will convert to GeoDataFrame.

```python
import geopandas as gpd
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.x, df.y))
```

```python
gdf.set_crs(epsg=3857, inplace=True)
```

```python
!pip install contextily
```

```python
import contextily as cx
import matplotlib.pyplot as plt
```

```python
from matplotlib import colors
cmap = colors.ListedColormap(['black', 'brown','gray'])
ax = gdf.plot(column='primary_fur_color', legend=True, edgecolor='black', cmap = cmap, linewidth=0.3)
cx.add_basemap(ax, crs = gdf.crs)
plt.show()
```

```python
#Creating a plot to see where the ages differ
ax = gdf.plot(column='age', legend=True, edgecolor='black', linewidth=0.3)
cx.add_basemap(ax)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
location_counts = gdf['location'].value_counts()
location_counts.plot(kind='bar')
plt.title('Distribution of Squirrel Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
